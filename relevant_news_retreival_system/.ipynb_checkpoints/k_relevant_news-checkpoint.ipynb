```python
def tokenize_and_count(df, columns):
    all_tokens = []
    for col in columns:
        tokens = df[col].dropna().str.split('[#;]').explode().str.strip()
        tokens = tokens[tokens.str.isalpha()]  # Remove numeric values
        all_tokens.extend(tokens)
    return Counter(all_tokens)

# Example usage
columns_to_tokenize = ['COUNTS', 'THEMES', 'LOCATIONS', 'PERSONS', 'ORGANIZATIONS']

token_counts = Counter()

for col in columns_to_tokenize:
    token_counts.update(tokenize_and_count(df, [col]))

sorted_token_counts = sorted(token_counts.items(), key=lambda x: x[1], reverse=True)
token_frequencies = np.array([count for token, count in sorted_token_counts])

# Calculate cumulative sum
cumulative_freq = np.cumsum(token_frequencies)
total_freq = cumulative_freq[-1]
optimal_threshold_idx = np.argmax(cumulative_freq >= 0.8 * total_freq)
```

```python
print(optimal_threshold_idx)
```

```python
selected_tokens = {token for token, count in sorted_token_counts[:optimal_threshold_idx + 1]}
```

```python
print(selected_tokens)
```

```python
def create_multihot_vector(row, selected_tokens, columns):
    tokens = set()
    for col in columns:
        if pd.notna(row[col]):
            col_tokens = set(re.split('[#;]', row[col]))
            tokens.update(col_tokens.intersection(selected_tokens))
    multihot_vector = [1 if token in tokens else 0 for token in selected_tokens]
    return multihot_vector

# Example usage
multihot_vectors = []
for index, row in df.iterrows():

    vector = create_multihot_vector(row, selected_tokens, columns_to_tokenize)
    multihot_vectors.append(vector)

bow = np.array(multihot_vectors)
```

```python
print(bow)
```

```python
X_Norm = preprocessing.normalize(bow)
```

```python
max_k=10
wcss = [] 
for k in range(2, max_k + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_Norm)
    wcss.append(kmeans.inertia_)
    print(f"Trained KMeans with k = {k}")

# Plotting the elbow curve
plt.plot(range(2, max_k + 1), wcss, marker='o')
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Inertia)')
plt.show()

# point where the curve bends through the maximum second derivative (get the maximum concativity)
diff_wcss = np.diff(wcss)
print(diff_wcss)
```

```python
optimal_k = np.argmax((diff_wcss==sorted(diff_wcss)[1])) + 3  #+2 bc starts with k=2, +1 bc sorted starts in 1
```

```python
optimal_k
```

```python
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X_Norm)
```

```python
tsne = TSNE(n_components=2, random_state=42,perplexity=10)
tsne_results = tsne.fit_transform(X_Norm)

# Plot the t-SNE results along with the cluster centroids
plt.figure(figsize=(10, 7))

# Plot the t-SNE results
plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.7)

# Plot the centroids
#cls_norm = preprocessing.normalize(kmeans.cluster_centers_)
#centroids_tsne = tsne.fit_transform(cls_norm)
#plt.scatter(centroids_tsne[:, 0], centroids_tsne[:, 1], s=300, c='red', label='Centroids')

plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE with K-Means Centroids')
plt.show()
```

```python
labels = kmeans.labels_
centroids = preprocessing.normalize(kmeans.cluster_centers_)
print(labels.shape)
print(centroids.shape)
```

```python
closest_samples_df = pd.DataFrame()

# For each cluster, find the 4 samples closest to the centroid
for cluster_idx in range(optimal_k):
    # Get all points in the cluster
    cluster_points_idx = np.where(labels == cluster_idx)[0]
    cluster_points = X_Norm[cluster_points_idx]

    # Calculate the Euclidean distance to the centroid
    distances = np.linalg.norm(cluster_points - centroids[cluster_idx], axis=1)

    # Get the indices of the 4 closest points
    closest_points_idx = cluster_points_idx[np.argsort(distances)[:4]]

    # Extract the rows from the original DataFrame and add a centroid label
    closest_samples = df.iloc[closest_points_idx].copy()
    closest_samples['centroid'] = cluster_idx

    # Append to the final DataFrame
    closest_samples_df = pd.concat([closest_samples_df, closest_samples])

# Reset index for the final DataFrame
closest_samples_df.reset_index(drop=True, inplace=True)
```
